This notebook is used to process the WSI files and store their dowscaled 256x256x3 sized images into .npy arrays
This notebook was run for multiple sessions and all the .npy files and their corresponding labels from all the sessions are stacked finally.

```python
start = 353
end = 450

repo_id = 8

j =18

DOWNLOAD_DIR = f"/kaggle/input/reghf{repo_id}"


# simply change the repo id and the "j" variable is just for filenames, 
# to download them all different sessions and stack all the .npy files and concat the json lists having labels
```

```python
import sys
sys.path.append("/kaggle/input/pip-openslide")
import openslide
```

```python
import psutil
import gc

def get_memory_usage():
    """Get current memory usage percentage"""
    return psutil.virtual_memory().percent

def cleanup_memory():
    """Clean up memory and show stats"""
    before = psutil.virtual_memory().percent
    gc.collect()
    after = psutil.virtual_memory().percent
    print(f"Memory cleanup: {before:.1f}% -> {after:.1f}%")
    return after

```

```python
import os
import time
import gc
from PIL import Image
import numpy as np
import openslide
from tqdm import tqdm


small , filenames = [] , []

start_time = time.time()

all_files = [f for f in os.listdir(DOWNLOAD_DIR) if f.lower().endswith('.tiff')]


for file in tqdm(all_files[start:end]):
    if time.time() - start_time >= ((11 * 3600)+(45*60)):
        print("11 hours elapsed, breaking loop")
        break

    slide = None  

    try:
        file_path = os.path.join(DOWNLOAD_DIR, file)
        slide = openslide.OpenSlide(file_path)
        filename = os.path.splitext(file)[0]

        image = slide.read_region((0, 0), 0, slide.level_dimensions[0]).convert("RGB")

        small_img = image.resize((256, 256), resample=Image.LANCZOS)
        small_np = np.transpose(np.array(small_img).astype(np.float32) / 255.0, (2, 0, 1))
        small.append(small_np)
        filenames.append(filename)

    except Exception as e:
        print(f"Error with {file}: {e}")

    finally:
        try:
            if slide is not None:
                slide.close()
            for var in ['image', 'small_img', 'large_img', 'small_np', 'large_np']:
                if var in locals():
                    del locals()[var]

            gc.collect()

            if 'cleanup_memory' in globals():
                cleanup_memory()

        except Exception as cleanup_error:
            print(f"Cleanup error: {cleanup_error}")
```

```python
try:
    np.save(f"small{j}.npy" , np.array(small))
    # np.save("large.npy" , np.array(large))
except Exception as e :
    print(e)
    # pass
```

```python
with open("/kaggle/input/codefiles/train.json" , "r") as f:
    d = json.load(f)

captions = []


for name in filenames:
    for x in d:
        if name in x["id"]:
            captions.append(x["report"])


with open(f"captions{j}.json", "w") as f:
    json.dump(captions , f , indent=2)



with open(f"filenames{j}.json", "w") as f:
    json.dump(filenames, f , indent=2)
```
